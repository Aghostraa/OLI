### Download parquet and load into Pandas df

````python
import pandas as pd

# download the decoded parquet file into a df
df = pd.read_parquet(oli.get_full_decoded_export_parquet())
df
````

````python
# look up all tags for one address
df.loc[df['address'] == '0x000000000004444c5dc75cB358380D2e3dE08A90']
````

````python
# look up one project
df.loc[(df['tag_id'] == 'owner_project') & (df['tag_value'] == 'uniswap')]
````

````python
# get all dexes on Arbitrum
df.loc[(df['chain_id'] == 'eip155:42161') & (df['tag_value'] == 'dex')]
````

````python
# pivot decoded_df by tag_id, tag_value and save into labels_df
df_pivot = df.pivot(index=['id', 'chain_id', 'address', 'attester', 'time_created', 'revoked', 'is_offchain'], columns='tag_id', values='tag_value')
df_pivot = df_pivot.reset_index()
df_pivot
````
